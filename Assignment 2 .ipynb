● Filling null values with a single value

```python
import pandas as pd
import numpy as np
df=pd.read_csv("StudentsPerformanceTest1.csv")
df
```

```python
ndf=df
ndf.fillna(0)
```

● Filling missing values using mean,median and standard deviation of that column.

```python
m_v=df["Math Score"].mean()
df['Math Score'].fillna(value=m_v,inplace=True)
df
```

```python
ndf.replace(to_replace=np.nan,value=-99)
```

```python
#delete null values usig dropna
ndf.dropna()
```

```python
#To Drop rows if all values in that row aremissing
ndf.dropna(how='all')
```

```python
#To Drop columns with at least 1 null value.
ndf.dropna(axis=1)
```

```python
#To drop rows with at least 1 null value inCSV file.making new data frame with dropped NA values
new_data=ndf.dropna(axis=0,how='any')
new_data
```
